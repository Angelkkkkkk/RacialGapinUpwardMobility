### upward transition probability (hereafter “UTP”)

#### 1880-1910

```python
df1880_1910_white_occ_unique = list(df1880_1910_white_pair_nona['occscore_1_y'].unique())
df1880_1910_black_occ_unique = list(df1880_1910_black_pair_nona['occscore_1_y'].unique())
```

```python
df1880_1910_utp_black = []
for i in df1880_1910_black_occ_unique:
    df = df1880_1910_black_pair_nona.loc[df1880_1910_black_pair_nona['occscore_1_y']<= i]
    df_upward = df.loc[df['occscore_2']>i]
    utp = df_upward.shape[0]/df.shape[0]
    df1880_1910_utp_black.append(utp)
```

```python
df1880_1910_utp_black = pandas.DataFrame({'father occ':df1880_1910_black_occ_unique,'utp_black':df1880_1910_utp_black})
df1880_1910_utp_black_sorted = df1880_1910_utp_black.sort_values(by=['father occ'])
```

```python
df1880_1910_utp_white = []
for i in df1880_1910_white_occ_unique:
    df = df1880_1910_white_pair_nona.loc[df1880_1910_white_pair_nona['occscore_1_y']<= i]
    df_upward = df.loc[df['occscore_2']>i]
    utp = df_upward.shape[0]/df.shape[0]
    df1880_1910_utp_white.append(utp)
```

```python
df1880_1910_utp_white = pandas.DataFrame({'father occ':df1880_1910_white_occ_unique,'utp_white':df1880_1910_utp_white})
df1880_1910_utp_white_sorted= df1880_1910_utp_white.sort_values(by=['father occ'])
```

```python
df1880_1910_utp_white_sorted.head()
```

```python
plt.plot(df1880_1910_utp_white_sorted['father occ'],df1880_1910_utp_white_sorted['utp_white'])
plt.plot(df1880_1910_utp_black_sorted['father occ'],df1880_1910_utp_black_sorted['utp_black'])
```

```python
ax = df1880_1910_UTP_sorted.plot(x='father occ',y='utp_white',kind = 'line',title='1880-1910 UTP By Racial Groups',ylabel='utp',figsize=(10,5))
df1880_1910_UTP_sorted.plot(ax = ax,x='father occ',y='utp_black',kind = 'line',grid = 'True')
```

```python
ax = df1880_1910_utp_white.plot(x='father occ', y='utp_white', style='o',figsize=(10,5),title='1880-1910 UTP',ylabel='utp')
df1880_1910_utp_black.plot(ax = ax, x='father occ', y='utp_black', style='o',grid = 'True')
```

So, lower income blacks are catching up in the upward mobility, the black-white upward transition probability difference for the bottom groups is shrinking over time. However, sadly, the upward transition probability difference for the middle and upper groups are increasing in the period of 1880-1920.
